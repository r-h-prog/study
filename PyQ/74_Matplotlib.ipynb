## オブジェクト指向グラフ

```python
# グラフの描画方法にはオブジェクト志向スタイルとMATLABスタイルがある
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

```python
# 描画オブジェクト(fig)とサブプロット(ax)を作成している
fig,ax = plt.subplots()
ax.plot([2,1,3])
ax.plot([1,3,2]);
```

```python
fig,axes = plt.subplots(2)
axes[0].plot([2,1,3])
axes[1].plot([1,3,2]);
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot([2,1,3])
axes[1].plot([1,3,2]);
```

## 画像ファイルの保存 - fig.savefig() -

```python
fig, ax = plt.subplots()
ax.plot([2,1,3])
fig.savefig('sample.png')
```

## 棒グラフ

```python
# データ
w = 0.4  # width
x1 = [0, 1, 2]
x2 = [w, 1 + w, 2 + w]
y1 = [1, 2, 3]
y2 = [3, 1, 2]
```

```python
fig, ax = plt.subplots()
ax.bar(x1, y1, width=w, color='r', alpha=0.5)
ax.bar(x1,y2, bottom=y1, width=w, color='b', alpha=0.5);
```

## 円グラフ

```python
fig, ax = plt.subplots()
# explode : 中心から離れる割合のリスト。今回は2番目の扇形を半径x0.1だけ話す
# autopct : パーセント表示用。「%」は%%と表す
# counterclock : 反時計回りかどうか
# ax.axis('equal') : 縦と横の比率を１にする。真円になる魔縁になる
ax.pie([2,1,3], explode=[0,0.1,0],
      labels=['Item 1', 'Item 2', 'Item 3'],
      autopct='%.1f%%', startangle=90,counterclock=False);
ax.axis('equal');
```

## ２軸のグラフ

```python
month = [4, 5, 6, 7, 8, 9]
sales_new = [11, 17, 12, 24, 35, 58] # 新規顧客の売上（千円）
sales_old = [102, 141, 127, 167, 202, 205] # 既存顧客の売上（千円）
rate_new = [round(new_ / (new_ + old) * 100)
            for new_, old in zip(sales_new, sales_old)]
labels = ['Sales & New rate', 'Sales(1000 yen)', 'New rate(%)']
```

```python
fig, ax1 = plt.subplots()
ax1.set_title(labels[0])
# ２軸のグラフを作成するにはtwinx()を使う
ax2 = ax1.twinx()
ax2.grid(False)

ax1.bar(month,sales_old, label='Old')
ax1.bar(month, sales_new, label='New',bottom=sales_old)
ax1.set_xlabel('Month')
ax1.set_ylabel(labels[1])
ax1.legend()

ax2.plot(month, rate_new, color='r')
ax2.set_ylabel(labels[2])
# ax2の下部を0にする
ax2.set_ylim(bottom=0);
```
